```python
all_datasets_path = '../data/input/final_standard_ncbi_input_species.csv'
train_file_path = '../data/339_output_genomes_ns/train_ncbi_bNone_1.csv'

import pandas

train = pandas.read_csv(train_file_path)

train.head()
```

```python
cutters = train.sequence.unique()
print(len(cutters))
cutters
```

```python
targets_covered_in_train = train.targets.unique()
print(len(targets_covered_in_train))
```

```python
test = pandas.read_csv(all_datasets_path)
print(test.shape)
test.head()
```

```python
test = test[~(test.species_name.isin(targets_covered_in_train))]
print(test.shape)
test
```

```python
import os
import sys
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)

from src.utils.guide_finder import GuideFinder

gf = GuideFinder()

# TODO: use a read mapping algo

paths: list[str] = list()
species_list: list[str] = list()
scores: list[int] = list()
strands: list[str] = list()
sequences: list[str] = list()
end_positions: list[int] = list()
start_positions: list[int] = list()
chromosomes_or_genes: list[str] = list()

for idx, row in test.iterrows():
    species = row['species_name']
    file_path = '../data/input/genomes/' + row['genome_file_name']

    for cutter in cutters:
        chrom_strand_start_end = gf.locate_guides_in_sequence(sequence=cutter, file_path=file_path)

        if chrom_strand_start_end:
            for tuple in chrom_strand_start_end:
                chromosome = tuple[0]
                strand = tuple[1]
                start_pos = tuple[2]
                end_pos = tuple[3]

                sequences.append(cutter)
                paths.append(row['genome_file_name'])
                scores.append(1)  # TODO fix for other than 1
                strands.append(strand)
                start_positions.append(start_pos)
                end_positions.append(end_pos)
                chromosomes_or_genes.append(chromosome)
                species_list.append(species)

pandas.DataFrame(list(zip(
        sequences,
        species_list,
        scores,
        chromosomes_or_genes,
        strands,
        start_positions,
        end_positions,
        paths,
        )),
    columns=['sequence','targets', 'score', 'chromosome_or_gene',
    'strand', 'start_position', 'end_position', 'path']).to_csv('train_test.csv', index=False)
            
```
