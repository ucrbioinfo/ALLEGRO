```python
import pandas as pd

df = pd.read_csv('interpro_results.csv')
df.fillna('N/A', inplace=True)

def shortify(s, go):
    while len(s) > 40:
        s = ' '.join(s.split(' ')[:-1]) + '...'
    
    if s == 'N/A':
        return 'N/A'
    else:
        return s + ' (' + go + ')'

df['biological_process_combined'] = df.apply(lambda row: ';'.join([f"{shortify(name, go)}" for go, name in zip(row['biological_process_go'].split(';'), row['biological_process_go_name'].split(';'))]), axis=1)
df['molecular_function_combined'] = df.apply(lambda row: ';'.join([f"{shortify(name, go)}" for go, name in zip(row['molecular_function_go'].split(';'), row['molecular_function_go_name'].split(';'))]), axis=1)
df['cellular_component_combined'] = df.apply(lambda row: ';'.join([f"{shortify(name, go)}" for go, name in zip(row['cellular_component_go'].split(';'), row['cellular_component_go_name'].split(';'))]), axis=1)

df['domains_combined'] = df.apply(lambda row: ';'.join([f"{shortify(name, go)}" for go, name in zip(row['domain_accessions'].split(';'), row['domains'].split(';'))]), axis=1)
df['families_combined'] = df.apply(lambda row: ';'.join([f"{shortify(name, go)}" for go, name in zip(row['family_accessions'].split(';'), row['families'].split(';'))]), axis=1)
df['superfamilies_combined'] = df.apply(lambda row: ';'.join([f"{shortify(name, go)}" for go, name in zip(row['superfamily_accessions'].split(';'), row['superfamilies'].split(';'))]), axis=1)
df['conserved_sites_combined'] = df.apply(lambda row: ';'.join([f"{shortify(name, go)}" for go, name in zip(row['conserved_site_accessions'].split(';'), row['conserved_sites'].split(';'))]), axis=1)
df['binding_sites_combined'] = df.apply(lambda row: ';'.join([f"{shortify(name, go)}" for go, name in zip(row['binding_site_accessions'].split(';'), row['binding_sites'].split(';'))]), axis=1)
df['active_sites_combined'] = df.apply(lambda row: ';'.join([f"{shortify(name, go)}" for go, name in zip(row['active_site_accessions'].split(';'), row['active_sites'].split(';'))]), axis=1)
df['repeats_combined'] = df.apply(lambda row: ';'.join([f"{shortify(name, go)}" for go, name in zip(row['repeat_accessions'].split(';'), row['repeats'].split(';'))]), axis=1)
df['other_features_combined'] = df.apply(lambda row: ';'.join([f"{shortify(name, go)}" for go, name in zip(row['other_feature_accessions'].split(';'), row['other_features'].split(';'))]), axis=1)

df
```

```python
def pretify_plt(plt):
    # Removing top and right spines
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['right'].set_visible(False)
    plt.gca().spines['left'].set_visible(False)
    plt.gca().spines['bottom'].set_visible(False)
    plt.tick_params(axis='both', which='both', length=0, labelcolor=cblack)
    plt.margins(x=0)
    plt.grid(alpha=0.3, linestyle='--')
    
    return plt

def set_xmargin(ax, left=0.0, right=0.3):
    import numpy as np

    ax.set_xmargin(0)
    ax.autoscale_view()
    lim = ax.get_xlim()
    delta = np.diff(lim)
    left = lim[0] - delta * left
    right = lim[1] + delta * right
    ax.set_xlim(left, right)

def disable_ax_spines(ax):
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['bottom'].set_visible(False)

cblack = '#262626'
cred = '#FE7070'
cgrey = '#ccc6c7'
cgreen = '#05B9A0'
cpurple = '#8B8EFF'
```

```python
from collections import Counter
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 600

def draw_plot(guide, df_column_go, xlabel, ylabel, title, size, color='#FFADAD'):
    go_list = list()

    for go_term in df[df['targeting_guide'] == guide][df_column_go].tolist():
        go_list.extend(list(set(go_term.split(';'))))

    data = Counter(go_list)

    sorted_data = dict(sorted(data.items(), key=lambda item: item[1], reverse=True))

    fig, ax = plt.subplots(figsize=size)
    bars = ax.barh(list(sorted_data.keys())[::-1], list(sorted_data.values())[::-1], color=color)

    # Adding the values next to each bar
    for bar in bars:
        xval = bar.get_width()
        ax.text(xval + 1, bar.get_y() + bar.get_height()/2, int(xval), ha='left', va='center')

    ax.set_xlabel(xlabel, color=cblack)
    ax.set_ylabel(ylabel, color=cblack)
    ax.set_title(title, color=cblack)
    ax.grid(alpha=0.3, linestyle='--')
    ax.tick_params(axis='both', which='both', length=0, labelcolor=cblack)
    disable_ax_spines(ax)
    plt.margins(0.05, 0)
    plt.tight_layout()
    plt.show()
```

```python
draw_plot('GAGGCTGGTATCTCCAAGGA',
          'domains_combined',
          'Counts',
          'Domain',
          'Protein Domain Counts for Targets of GAGGCTGGTATCTCCAAGGA',
          (8, 4),
          color=cpurple)
```
